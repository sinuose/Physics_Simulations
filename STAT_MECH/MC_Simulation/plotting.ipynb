```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = np.genfromtxt('out.csv', delimiter=',')
energy, magnetization = zip(*data[1:])
energy = np.array(energy)
magnetization = np.array(magnetization)

exp_energy = energy.reshape(len(energy)//100, 100).mean(axis=1)
exp_magnetization = magnetization.reshape(len(magnetization)//100, 100).mean(axis=1)

time = np.arange(len(energy))

```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize = (12,6), dpi=300)
ax1.set_title("Lattice Energy vs. MCS steps")
ax1.plot(time, energy/32**2, 'ro--', markersize=2,label="Lattice Energy")
ax1.set_xlabel("Time [MCS Step]")
ax1.set_ylabel("Energy $[1/J]$")

ax1.grid()
ax1.legend()

ax2.set_title("Lattice Magnetization: $1/L^2\sum \sigma$ vs. MCS steps")
#ax2.plot(time, energy/(2*32**2), 'ro--', markersize=2,label="Lattice Energy")
ax2.plot(time, magnetization/32**2, 'bo--', markersize=2, label="Lattice Magnetization")
ax2.set_xlabel("Time [MCS Step]")
ax2.set_ylabel("Magnetization")

ax2.grid()
ax2.legend()
plt.show()
```
