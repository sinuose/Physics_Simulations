```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def T(x, a, q):
    return 3*a*q*x/2 * (np.log(x+1) - np.log(1-x))** (-1)

def T_sp(x, a, q):
        return 3*a*q*(1-x**2)/4 
```

```python
a = 1
q = 1

x_range = np.arange(-1, 1, 0.001)
y_range = T(x_range, a, q)
sp_range = T_sp(x_range, a, q)
```

```python
plt.figure(figsize=(12,8))
plt.plot(x_range, y_range, 'ro-', label = "Coexistence Line")
plt.plot(x_range, sp_range, 'b--', label = "Spinodal Line")
plt.plot(0, sp_range[len(sp_range)//2], 'yo', label = "Critical Temperature", markersize=10)
plt.title("Phase Diagram of Binary Alloy", fontsize=20)
plt.text(-0.15, 0.5, "2-Phase ", fontsize=20)
plt.text(-0.15, 0.9, "1-Phase ", fontsize=20)
plt.legend()
plt.xlabel("x", fontsize=20)
plt.ylabel("T", fontsize=20)
plt.ylim(0, 1)
plt.grid()
plt.show()
```
